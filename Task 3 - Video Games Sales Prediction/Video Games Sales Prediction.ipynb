### 3. Exploratory Data Analysis and Data Preprocessing

The data is subjected to multiple visualizations and methods to find and handle the missing values and outliers if there are any. This step is crucial, as this one helps to find the trends and relationships between variables. 

```python
data.isnull().sum()
```

Since there are some columns which are unneccessary for the scope of this project, we can drop it.

```python
data.drop(columns = ['Year_of_Release', 'Developer', 'Publisher', 'Platform'], inplace = True)
print('Feature name || Total missing values')
print(data.isnull().sum())
```

```python
data.head()
```

```python
print(data['User_Score'].dtype)
```

Here, there are some columns with the object datatype. We need to make sure that these columns should be converted into  numerical otherwise it will affect the working of our model.

```python
for i in range(len(data['User_Score'])):
    if type(data['User_Score'].iloc[i]) == type("String") and data['User_Score'].iloc[i] != 'tbd':
        data.loc[i, 'User_Score'] = float(data['User_Score'].iloc[i])
    if data['User_Score'].iloc[i] == 'tbd':
        data.loc[i, 'User_Score'] = 0.0
```

The missing values have been imputed with the mean values as follows

```python
imputer = SimpleImputer(strategy = 'mean')
data.iloc[:,[7,8,9,10]] = imputer.fit_transform(data.iloc[:,[7,8,9,10]])
data.head()
```

```python
data.isnull().sum()
```

Since the columns 'Name', 'Genre' and 'Rating' are categorical features, the missing values should be handles and they should be encoded into numerical variables 

```python
imputer = SimpleImputer(strategy = 'constant', fill_value = 'NA')
data.iloc[:, [0,1,11]] = imputer.fit_transform(data.iloc[:, [0,1,11]])
```

The Genre column has been encoded numerically according to the category as follows

```python
data.replace({'Genre':{'Sports' : 0, 'Platform' : 1, 'Racing' : 2, 'Role-Playing' : 3, 'Puzzle' : 4, 'Misc' : 5, 'Shooter' : 6, 'Simulation' : 7, 'Action' : 8, 'Fighting' : 9, 'Adventure' : 10, 'Strategy' : 11, 'NA' : 12}}, inplace = True)
```

```python
X = data.iloc[:, [0, 1, 2, 3, 4, 5, 7, 8, 9, 10]]
y = data['Global_Sales']
X.head()
```

The below code makes sure that the datatype of the user_score column is of datatype int32

```python
X['User_Score'] = pd.to_numeric(X['User_Score'], errors='coerce')
X['User_Score'] = X['User_Score'].fillna(0).astype(int)  #Avoid the warnings
```

```python
print(X['User_Score'].dtype)
```

YAY! Finally, our dataset has been cleaned properly for efficient training

### 4. Data Splitting

The dataset has been splitted into training and testing sets which accounts for 80% of data for training and 20% for testing

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

The game names are stored separately as follows. This is to access the game names finally. And then the train and test sets are updated as follows

```python
games_train = x_train.iloc[:, [0]]
games_test = x_test.iloc[:, [0]]
games_train
```

```python
X_train = x_train.iloc[:, 1:] 
X_test = x_test.iloc[:, 1:]
```

```python
X_train.head()
```

### 5. Model Training

For this project, I've used the XGB Regressor model for the accurate predictions

### Introduction to XGB Regressor Model

The XGBoost (Extreme Gradient Boosting) Regressor model is a powerful supervised machine learning algorithm used for regression tasks. It is an enhanced version of the gradient boosting algorithm that combines multiple weak prediction models, typically decision trees, to create a strong predictive model.

XGBoost is known for its efficiency, scalability, and accuracy. It excels in handling complex, high-dimensional datasets and can capture non-linear relationships between features and target variables. The algorithm follows a boosting framework where each subsequent model is trained to correct the mistakes made by the previous models.



```python
from xgboost import XGBRegressor
model = XGBRegressor(n_estimators = 200, learning_rate = 0.08)
model.fit(X_train, y_train)
```

```python
# Predicting test set results
y_pred = model.predict(X_test)
print(y_pred)
```

### 6. Prediction and Evaluation 
The trained model is further tested for its accuracy and then evaluated using metrics such as R2 score and MSE

```python
y_pred = y_pred.reshape(-1,1)
y_test = y_test.values.reshape(-1,1)
predictions = np.concatenate([games_test, y_pred, y_test], axis = 1)
predictions = pd.DataFrame(predictions, columns = ['Name', 'Predicted_Global_Sales', 'Actual_Global_Sales'])
```

```python
predictions.head(20)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
import math
r2_score = r2_score(y_test, y_pred)
rmse = math.sqrt(mean_squared_error(y_test, y_pred))

print(f"r2 score of the model : {r2_score:.3f}")
print(f"Root Mean Squared Error of the model : {rmse:.3f}")
```
